This article is written by Ira Serke, a property agent from Berkley California. The article discusses effective techniques used by Ira Serke and his team that helped them win maximum listings and in turn generated sale proceeds. The article also covers daily habits of an effective property agent.
Ira Serkey, started his journey by shifting his focus from the amount of property sales (from being in top 5% of property agents by transactions in the Berkley area) to a happier life by having more than enough business. ## Opinion
1. The article describes the techniques of driving referrals and listings by Ira Serke and his team which, is very interesting and practical. Author also discusses source of business in the form of referrals. The idea of an agent giving away business to another agent is new approach to real estate business. 
2. Ira correctly describes the success of maximum listing in various techniques such as preparing a presentation for the seller to answer any question they may have before they ask them; or personal involvement of Ira in the activities conducted by the local college continue to provide the sustained involvement, introductions. Ira associated himself with a nationwide network of agents for this purpose.  Networking is achieved by investing in local products and services.

3. Client referrals are mostly driven by hosting celebrations that are popular and attract large crowds. Other technique is Agent Referrals which generates good will among the agents population and give Ira extra income in the form of the 10% referral fee they make on each such deal they pass on. 
4. Ira successfully made maximum money by spending least amount of time. Therefore spending 3 minutes on the phone call with the seller and then forwarding the listing to the other agent seems to be a much better deal to Ira.

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Sold as Lising price', 'Above Listing price', 'Below Listing price', 'Not sold'
sizes = [30, 40, 15, 5]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
x = np.linspace(500, 800, 100000)

plt.plot(x, x, label='zEstimate')
plt.plot(x, x*0.95, label='Actual Sales price')
#plt.plot(x, x**3, label='cubic')

plt.xlabel('Zestimate prices')
plt.ylabel('Actual sales prices')

plt.title("Comparison graph of Zillow Estimates and Atcual prices of Homes")

plt.legend()

plt.show()
```
